```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


'''downlaod haberman.csv'''
#Load Haberman.csv into a pandas dataFrame.
cancer_patient = pd.read_csv("haberman.csv")

```

```python
#from above output we can see that haberman.csv file has 306 instances of data and 4 attributes.
```

```python
#(Q) What are the column names in our dataset?
print (cancer_patient.columns)
```

```python
# above output gives the column names of our data set.
```

```python
#(Q) How many data points for each class are present? 
#  1 = the patient survived 5 years or longer 2 = the patient died within 5 year
cancer_patient["status"].value_counts()
```

```python
#above data shows us that 225 patients survived 5 years or longer while 81 patients died within 5 years.
```

```python
cancer_patient.plot(kind='scatter', x='age', y='nodes') ;
plt.show()
```

```python
#From the above 2-D scatter plot we are not able to conclude anything as most of the points are overlapping on each other and are in same colour.
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(cancer_patient, hue="status", size=5) \
   .map(plt.scatter, "age", "nodes") \
   .add_legend();
plt.show();
```

```python
#from above plot blue dots shows patients survived for 5 years or more while orange dots shows patient died before 5 years, but
# since still data points are overlapping each others we are not able to conclude anything clearly.
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(cancer_patient, hue="status", size=3, vars=['age','year','nodes']);
plt.show()
```

```python
#in the above pair plots we would be comsidering plot no.2,3 & 6, but Since in plot no.2 & 6 most of the data sets are
#overlapping , so we are not able to conclude anything and will not be considering them.
#plot no.3 has some distingushiable points and is better then other 2 plots,hence we will be using this plot attributes age 
#& nodes .
```

```python
import numpy as np
cancer_patient_alive = cancer_patient.loc[cancer_patient["status"] == 1];
cancer_patient_dead = cancer_patient.loc[cancer_patient["status"] == 2];


plt.plot(cancer_patient_alive["nodes"], np.zeros_like(cancer_patient_alive['nodes']), 'o')
plt.plot(cancer_patient_dead["nodes"], np.zeros_like(cancer_patient_dead['nodes']), 'o')

plt.show()
```

```python
#from the above plot we can see that the data set of patients died within 5 years are overlapping on the data set of patients 
#survived for 5 years or longer, so we need to make further analysis for better understanding.
```

```python
sns.FacetGrid(cancer_patient, hue="status", size=5) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```

```python
#From the above PDF of nodes we can see that patient with less than 5 nodes affected has more survival chances while that of
#patient with more than 5 nodes affected has less survival chances for longer period.
```

```python
sns.FacetGrid(cancer_patient, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

```python
#As in above PDF of age we can see that most of the datapoints are overlapping between the age range of 30 to 75, hence we are
#not able to conclude anything from this plot.
```

```python
sns.FacetGrid(cancer_patient, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.show();
```

```python
#here also most of the datasets are overlapping on each other making it difficult to conclude anything.
```

```python
counts, bin_edges = np.histogram(cancer_patient_alive['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(cancer_patient_alive['nodes'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();
```

```python
#From the above plot we can conclude that 85% of patients will have long survival chances if their nodes affected is less than 
#5
```

```python
counts, bin_edges = np.histogram(cancer_patient_alive['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(cancer_patient_dead['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)

plt.show();
```

```python
# From the above plot 55% of patients with nodes less than 10 will survive less than 5 Years
```

```python
print("Means:")
print(np.mean(cancer_patient_alive["nodes"]))
#Mean with an outlier.
print(np.mean(np.append(cancer_patient_alive["nodes"],50)));
print(np.mean(cancer_patient_dead["nodes"]))


print("\nStd-dev:");
print(np.std(cancer_patient_alive["nodes"]))
print(np.std(cancer_patient_dead["nodes"]))
```
